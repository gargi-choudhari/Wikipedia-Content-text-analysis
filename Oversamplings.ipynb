```python
import pandas as pd
from sklearn.utils import resample
df = pd.read_excel(r"C:\Users\gargi\Downloads\Logit model data 1.xlsx")
df
```

```python
import wikipedia
names = df['Name']
i = 0
for name in names:
    content = wikipedia.page(name, auto_suggest=False).content
    df.at[i,'Content'] = content
    i= i+1
```

```python
df
```

```python
df_male = df[df['Gender'] == 'Male']
df_female = df[df['Gender'] == 'Female']
```

```python
df.to_csv('upsampled final.csv')
```

```python
df_female_upsampled = resample(df_female, 
                               replace=True,
                               n_samples=len(df_male), 
                               random_state=123) 
```

```python
df_upsampled = pd.concat([df_male, df_female_upsampled])
```
