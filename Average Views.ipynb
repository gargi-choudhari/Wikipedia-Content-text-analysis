```python
import requests
import json
from pprint import pprint
import pandas as pd
df = pd.read_excel('Logit model data.xlsx')
headers = {
    'User-Agent': 'pyMediaWiki-User-Agent-String',
}
total_views = 0
for name in df['Name'].tolist():
  name = name.replace(' ', '_')
  name = name.strip()
  resp = requests.get(f'https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/en.wikipedia/all-access/all-agents/{name}/daily/2015100100/2023090100', headers=headers)
  if resp.status_code:
    data = resp.json()
    views = 0
    for i in range(len(data['items'])):
      views = views + data['items'][i]['views']

    avg_views = float(views/len(data['items']))
    total_views+=avg_views
print(total_views)
```
