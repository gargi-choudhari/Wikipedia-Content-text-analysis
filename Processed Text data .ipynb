```python
#punctuation removal 
import string
import re

def remove_punctuation(input_string):
    # Make a regular expression that matches punctuations
    punct = re.compile('[%s]' % re.escape(string.punctuation))
    
    # Use the sub function to replace the matched string with ''
    return punct.sub('', input_string)

df['Content'] = df['Content'].apply(remove_punctuation)

#lower case 
df['Content'] = df['Content'].str.lower()

#symbols and numbers(special characters)
import re
def remove_symbols_numbers(text):
    return re.sub('[^a-zA-Z]', ' ', str(text))

df['Content'] = df['Content'].apply(remove_symbols_numbers)

#Stopwords english removal
import nltk
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    return ' '.join([word for word in str(text).split() if word.lower() not in stop_words])

df['Content'] = df['Content'].apply(remove_stopwords)
df
```

```python
#proper noun removal
import spacy

nlp = spacy.load('en_core_web_sm')


def remove_proper_nouns(text):
    doc = nlp(text)
    tokens = [token.text for token in doc if token.pos_ != 'PROPN']
    return " ".join(tokens)

df['Content'] = df['Content'].apply(remove_proper_nouns)
df
```

```python
#lemmatization 
import spacy

nlp = spacy.load('en_core_web_sm')

def lemmatize_text(text):
    doc = nlp(text)
    return " ".join([token.lemma_ for token in doc])

df['Content'] = df['Content'].apply(lemmatize_text)
```
